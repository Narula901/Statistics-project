```python
# Import Libraries
import pandas as pd
import numpy as np 
```

```python
# Fetch Data from Excel
Employee_data = pd.read_csv('C:\\Users\\JITENDRA\\Downloads\\Data for Stat Projects\\employee_data.csv')
Employee_data.info()
```

```python
# Creating function to Convert Numerical data (Healthy_eating) into Categorical (Diet)

def condition(x):
    if x<3:
        return "Bad Diet"
    elif x>=3 and x<=6:
        return "Average Diet"
    else:
        return "Superb Diet"


Employee_data['Diet'] = Employee_data['healthy_eating'].apply(condition)
print(Employee_data)
```

```python
# create Summary for Groups and Diet 
#  We can find out the Conditional Probability from this Summary

ctab_JointP = pd.crosstab(Employee_data.groups, Employee_data.Diet, margins= True, margins_name='Total')
ctab_JointP

```

```python
# Probability of Above Summary 
# These are Joint Probabilites

ctab = pd.crosstab(Employee_data.groups, Employee_data.Diet, margins= True, margins_name='Total', normalize = True )
ctab
```

```python
# Conditional Probability based on Columns wise

ctab_conditionalP = pd.crosstab(Employee_data.groups, Employee_data.Diet, margins= True, margins_name='Total', normalize = 'columns')
ctab_conditionalP
```

```python
# Conditional Probability based on Index/Rows wise

ctab_conditionalP = pd.crosstab(Employee_data.groups, Employee_data.Diet, margins= True, margins_name='Total', normalize ='index')
ctab_conditionalP
```

```python
# Crosstable for Groups and Diet on the basis of Salary with different aggergate Function

ctab_salary = pd.crosstab(Employee_data.groups, Employee_data.Diet, values= Employee_data.salary, aggfunc={min,max,np.mean}).round(0)
ctab_salary
```
